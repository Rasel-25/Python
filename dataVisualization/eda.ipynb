# line plot

```python
x=np.arange(0,100)
y=np.arange(0,100)
```

```python
plt.figure(figsize=(20,10))  # Create a new figure, or activate an existing figure.
plt.plot(x,y)
plt.title('X vs Y')
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.show() #Display all open figures.
```
