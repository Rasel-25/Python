# line plot

```python
x=np.arange(0,100)
y=np.arange(0,100)
```

```python
# Straight line
plt.figure(figsize=(16,8))  # Create a new figure, or activate an existing figure.
plt.plot(x,y, label="straight line",linewidth=1, marker='^')
plt.title('X vs Y')
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.legend() # recognize the label
plt.grid(True)
plt.xticks(np.arange(0,100,5)) # Get or set the current tick locations and labels of the x-axis. 
plt.yticks(np.arange(0,100,5))
plt.show() #Display all open figures.
```

```python
# Sine & cos curve
x=np.arange(0,15)
y=np.sin(x)
plt.figure(figsize=(10,5))  # Create a new figure, or activate an existing figure.
plt.plot(x,y, label="sin curve",linewidth=1, marker='*')
plt.plot(x,np.cos(x), label="Cosine curve",linewidth=1, marker='*')

plt.title('X vs Y')
plt.xlabel("X-Axis")
plt.ylabel("sin(x)")
plt.legend() # recognize the label
plt.grid(True)
plt.show() #Display all open figures.
```

```python
# Bar chart show the frequencies of discrete variables
Age=[10,15, 25, 20, 30]
No_of_people=[20,30,25,15,10]
plt.figure(figsize=(10,5))  # Create a new figure, or activate an existing figure.
plt.bar(Age,No_of_people, label="Age data")

plt.title('X vs Y')
plt.xlabel("Age")
plt.ylabel("People")
plt.legend() # recognize the label
plt.grid(True)
plt.show() #Display all open figures.
```

# Horijontal Bar

```python
# Bar chart
langs=['Pytho','JAVA','C','C++']
rating=[87,68,75,80]
plt.figure(figsize=(10,5)) 
plt.barh(langs,rating, label="Programming Language")

plt.title('X vs Y')
rating=[87,68,75,80]
plt.xlabel("rating")
plt.ylabel("langs")
plt.legend() # recognize the label
plt.grid(True)
plt.show() #Display all open figures.
```

# Histogram

```python
# Bar chart show the frequencies of continuous variables
h=np.random.randn(300)

plt.figure(figsize=(10,5)) 
plt.hist(h,bins=10, label='Random distribution') # bins specify the number of bar
plt.title('Histogram')

plt.xlabel("range")
plt.ylabel("Frequencies")
plt.legend() # recognize the label
plt.grid(True)
plt.show() #Display all open figures.
```

# Pie Chart

autopct to display the percent value using Python string formatting. For example, autopct='%1.1f%%'

explode show any cell uniquely 

```python
sports = ['Football', 'Cricket','Hocky','Basket Ball',]
participation=[380,350,150,120]
plt.figure(figsize=(10,5))
plt.pie(participation, labels=sports, autopct='%1.1f%%',explode=(0,.1,0,0),shadow=(0,0,.5,0)) 
plt.show()
```

# Scatter Plot

```python
x=np.random.random(200)
y=np.random.random(200)
plt.figure(figsize=(10,10))  
plt.scatter(x,y, label="Random data", color='darkorange',edgecolors='green')

plt.title('X vs Y')
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.legend() 
plt.grid(True)
plt.show()
```

# 3D plots

 3D scatter plot is created by using ax.scatter3D() the function of the matplotlib library which accepts a data sets of X, Y and Z to create the plot while the rest of the attributes of the function are the same as that of two dimensional scatter plot.

```python
x=np.random.random(200)
y=np.random.random(200)
z=np.random.random(200)
plt.figure(figsize=(15,10))
axes=plt.axes(projection='3d') #Add an axes to the current figure and make it the current axes.
axes.scatter3D(x,y,z)  
axes.set_xlabel('x_axis')
axes.set_ylabel('y_axis')
axes.set_zlabel('z_axis')
plt.show()
```

```python
# Creating dataset
z = np.random.randint(100, size =(50))
x = np.random.randint(80, size =(50))
y = np.random.randint(60, size =(50))
 
# Creating figure
fig = plt.figure(figsize = (10, 7))
ax = plt.axes(projection ="3d")
 
# Creating plot
ax.scatter3D(x, y, z, color = "green")
plt.title("simple 3D scatter plot")
 
# show plot
plt.show()
```
